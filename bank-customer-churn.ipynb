```python
# One-Hot encoding the data
data = pd.get_dummies(data, columns=['Gender', 'Geography'])
data.head()
```

```python
from sklearn.model_selection import train_test_split
y=data.Exited
x=data.drop(['Exited'], axis='columns')
x.head()
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_predict = classifier.predict(x_test)
```

```python
#Evaluate our model
from sklearn.metrics import accuracy_score
print("Accuracy of Logistic Regression before Normalizing data: ", accuracy_score(y_test, y_predict))
```

```python
# Normalize data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#transform training dataset
x_train_norm = scaler.fit_transform(x_train)
x_test_norm = scaler.transform(x_test)
```

```python
classifier_norm = LogisticRegression(random_state = 0)
classifier_norm.fit(x_train_norm, y_train)
```

```python
y_predict_norm = classifier_norm.predict(x_test_norm)
print ("Accuracy of Logistic Regression after Normalizing data :", accuracy_score(y_test, y_predict_norm))
```

<h1> Accuracy increased by about 2% by just normalizing the data </h1>
<h2> Accuracy before normalization: 0.78 </h2>
<h2> Accuracy after normalization: 0.80 </h2>

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_predict_rf = classifier.predict(x_test)
print(y_predict_rf)
```

```python
print("Accuracy of Random Forest Classifier before data normalizing: ", accuracy_score(y_test, y_predict_rf))
```

```python
classifier_norm = RandomForestClassifier(n_estimators=100, random_state=0)
classifier_norm.fit(x_train_norm, y_train)
```

```python
y_predict_rm_norm = classifier_norm.predict(x_test_norm)
print("Accuracy of Random Forest Classifier after data normalization: ", accuracy_score(y_test, y_predict_rm_norm))
```
